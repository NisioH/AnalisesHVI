```python
import pandas as pd
import plotly.express as px
```

```python
df = pd.read_excel('Planilha Orçamento 24-25.xlsx', sheet_name='Revisões')
df.drop([0], inplace=True)
df.head()
```

```python
# Iterando sobre cada valor único na coluna "Discriminação"
for discriminacao in df["Discriminação"].unique():
    # Filtrando os dados para o gráfico atual
    df_filtrado = df[df["Discriminação"] == discriminacao]

    # Transpondo os meses e valores para o formato do gráfico
    df_grafico = df_filtrado.melt(id_vars=["Discriminação"], 
                                  var_name="Meses", 
                                  value_name="Valores")
    
    df_grafico = df_grafico[~df_grafico["Meses"].isin(["VALOR TOTAL", "Quantidade"])]
    


    df_grafico["Valores"] = df_grafico["Valores"].apply(lambda x: f"R${x:,.2f}")



    # Criando o gráfico de linha
    fig = px.line(df_grafico, x="Meses", y="Valores", 
                  title=f"Evolução de Valores: {discriminacao}")

    # Exibindo o gráfico
    fig.show()
```
